Considere um sistema linear e invatriante no tempo com condições inicias $y[-1] = 1$ e $y[-2] = 1$ e descrito pela equação de diferença:

$$
y[n] - y[n-1] - 0.21 y[n-2] = x[n]
$$

e sinal de entrada $x[n] = (1-0.9^n)u[n]$

\begin{align*}
Y[z] (1 - z^{-1} &- 0.21 z^{-2}) = X[z] \\

\frac{X[z]}{Y[z]} &= 1 - z^{-1} - 0.21 z^{-2} \\
\frac{Y[z]}{X[z]} &= \frac{1}{1 - z^{-1} - 0.21 z^{-2}} \\
\frac{Y[z]}{X[z]} &= \frac{z^2}{z^2 - z - 0.21} \\
\frac{Y[z]}{X[z]} &= \frac{z^2}{(z - 0.3)(z - 0.7)} \\
\frac{Y[z]}{zX[z]} &= \frac{z}{(z - 0.3)(z - 0.7)} \\
\frac{Y[z]}{zX[z]} &= \frac{A}{z - 0.3} + \frac{B}{z - 0.7} \\
A &= \left[(z-0.3) \frac{z}{(z - 0.3)(z - 0.7)}\right]_{z=0.3} = \frac{0.3}{0.3 - 0.7} = -3/4 \\
B &= \left[(z-0.7) \frac{z}{(z - 0.3)(z - 0.7)}\right]_{z=0.7} = \frac{0.7}{0.7 - 0.3} = 7/4 \\
\frac{Y[z]}{zX[z]} &= \frac{-3/4}{z - 0.3} + \frac{7/4}{z - 0.7} \\
H[z] = \frac{Y[z]}{X[z]} &= \frac{-3}{4} \frac{z}{z - 0.3} + \frac{7}{4} \frac{z}{z - 0.7} \\
\end{align*}

Portanto, a função de transferência é dada por:

$$
H[z] = \frac{-3}{4} \frac{z}{z - 0.3} + \frac{7}{4} \frac{z}{z - 0.7} \xLeftrightarrow{\mathcal{Z}} h[n] = \frac{-3}{4} (0.3)^n u[n] + \frac{7}{4} (0.7)^n u[n] \\
$$
